```python
# Start training
# Load a pretrained model (weights will download automatically)
model = YOLO('custom_training/content/runs/detect/yolo11tr7/weights/best.pt')

results = model.train(
    data='/content/custom_training/dataset8/data.yaml',
    epochs=32,      # Number of rounds to train
    imgsz=640,      # Image size
    batch=32,       # Adjust based on GPU memory
    name='yolo11tr8',
    conf=0.4,
    iou=0.45
)
# Zip the results and move to Drive
!zip -r /content/yolo11tr8.zip /content/runs/detect/yolo11tr8
!cp /content/yolo11tr8.zip /content/drive/MyDrive/yolo_models
```
